```python
def Gaus(x,mean,sigma,norm=False):
        if (sigma == 0): return 1.e30
        arg = (x-mean)/sigma
        res = np.exp(-0.5*arg*arg)
        if not norm: return res
        return res/(np.sqrt(2*np.pi)*sigma)

def CedarQE_R9880U_110(wavelength):
        par = [1489.053765000671,     -20.61340505642701,     0.09607362916193821,
                      -0.000144918944048782, -1.087924475686453e-07, 3.619104979507752e-10,
                      2.742092765095943e-13, -1.067200613381487e-15, 6.333980140159196e-19,
                      4.675391577876988,     505.1903283978535,      15.37334879108591,
                      -23.08738129086531,    358.7521218115685,      53.63424346389683]

        x = wavelength
        x1 = x if x < 650 else 650
        qe = 0
        for i in range(9):
            qe += par[i] * pow(x1, i);
        qe += par[9] * Gaus(x1, par[10], par[11])
        qe += par[12] * Gaus(x1, par[13], par[14])
        qe *= 0.01
        if(x > 650):
            qe *= (1 - (x - 650) / (675 - 650))
        if(qe < 0 or x < 200):
            qe = 0
        return qe
    
def CedarQE_R7400U_03(wavelength):
        par = [-58.41145755814,     1.450540667766,      -0.01561331198442,
                                    9.545010080831e-05,  -3.648461145542e-07, 9.047599515597e-10,
                                    -1.457151808585e-12, 1.471328774241e-15,  -8.46121819724e-19,
                                    2.11384701372e-22]

        x = wavelength
        if(x < 180):
            return 0
        if(x > 660):
            return 0
        qe = 0;
        for i in range(10):
            qe += par[i] * pow(x, i);
        if(qe < 0):
            qe = 0
        return qe



def Cedar_Eff_QE(wavelength):
    if (random.random() < 32/48):
        return CedarQE_R9880U_110(wavelength)
    else:
        return CedarQE_R7400U_03(wavelength)

    
def CedarWQuartzWindowAndFilterTransmittance(wavelength):
    x = wavelength
    f = 0
    if(x > 350):
        f = 0.905 + 0.00008 * (x - 350)
    elif(x > 300):
        f = 0.875 + 0.030 * (x - 300) / 50.0;
    elif(x > 280):
        f = 0.845 + 0.030 * (x - 280) / 20.0;
    elif(x > 270):
        f = 0.815 + 0.030 * (x - 270) / 10.0;
    elif(x > 260):
        f = 0.760 + 0.055 * (x - 260) / 10.0;
    elif(x > 250):
        f = 0.670 + 0.090 * (x - 250) / 10.0;
    elif(x > 240):
        f = 0.490 + 0.180 * (x - 240) / 10.0;
    else:
        f = 0.232 + 0.258 * (x - 230) / 10.0;
    if(f < 0.001):
        f = 0.001
    if(f > 0.999):
        f = 0.999
    return f
```

```python
    
fposition = inFilek['SlimMC;1']['Cedar/TDetectorVEvent/fHits/fHits.fPosition'].array()
fenergy = inFilek['SlimMC;1']['Cedar/TDetectorVEvent/fHits/fHits.fEnergy'].array()

fPDGCode = inFilek['SlimMC;1']['Generated/fKineParts/fKineParts.fPDGcode'].array()

fposition_x = list(fposition['fX'])
fposition_y = list(fposition['fY'])
fposition_z = list(fposition['fZ'])



x_e = []
y_e = []
z_e = []

r_e = []
w_e = []

r_e_K = []
w_e_K = []

r_e_pro = []
w_e_pro = []

r_e_pi = []
w_e_pi = []

for i in range(len(fposition_x)):
    if (i % 100 == 0):
        print(i)
    for t in range(len(fposition_x[i])):
        x = fposition_x[i][t]
        y = fposition_y[i][t]
        r = np.sqrt(x**2 + y**2)

        #print(r)

        w = ELConvert(fenergy[i][t])*1e-3
        if (random.random() > CedarWQuartzWindowAndFilterTransmittance(w)):
        	continue
     
        if(random.random() > Cedar_Eff_QE(w)):
            continue
        
        x_e.append(x)
        y_e.append(y)
        w_e.append(w)
        r_e.append(r)
        #print(fPDGCode[i][0])
        if fPDGCode[i][0] == 2212:
            #print(r_e_pro)
            w_e_pro.append(w)
            r_e_pro.append(r)
        elif fPDGCode[i][0] == 321:
            w_e_K.append(w)
            r_e_K.append(r)  
        elif fPDGCode[i][0] == 211:
            w_e_pi.append(w)
            r_e_pi.append(r) 
        else:
            pass
```
